1. Fórmulas de Conversión

Para movernos entre las escalas Celsius y Fahrenheit, utilizamos relaciones lineales basadas en los puntos de congelación y ebullición del agua.

De Celsius a Fahrenheit:

$$°F = (°C \times \frac{9}{5}) + 32$$

(También puedes usar $1.8$ en lugar de $\frac{9}{5}$ para facilitar el cálculo decimal).De Fahrenheit a Celsius:$$°C = \frac{(°F - 32) \times 5}{9}$$

(O bien: $°C = \frac{°F - 32}{1.8}$).

2. Ejemplos Resueltos

Ejemplo A: Convertir 25°C a Fahrenheit
1. Identificar datos: Tenemos $°C = 25$.
2. Sustituir en la fórmula: $°F = (25 \times 1.8) + 32$.
3. Multiplicar: $25 \times 1.8 = 45$.
4. Sumar 32: $45 + 32 = 77$.
* Resultado: 25°C equivalen a 77°F.

Ejemplo B: Convertir 95°F a Celsius

1. Identificar datos: Tenemos $°F = 95$.
2. Sustituir en la fórmula: $°C = (95 - 32) \div 1.8$.
3. Restar 32: $95 - 32 = 63$.
4. Dividir entre 1.8: $63 \div 1.8 = 35$.
* Resultado: 95°F equivalen a 35°C.

```python
def convertirCF():
    """Convierte de Celsius a Fahrenheit"""
    celsius = float(input("\nIngresa los grados Celsius: "))
    fahrenheit = (celsius * 9/5) + 32
    print(f"--- {celsius}°C equivalen a {fahrenheit:.2f}°F ---")

def convertirFC():
    """Convierte de Fahrenheit a Celsius"""
    fahrenheit = float(input("\nIngresa los grados Fahrenheit: "))
    celsius = (fahrenheit - 32) * 5/9
    print(f"--- {fahrenheit}°F equivalen a {celsius:.2f}°C ---")

def menu():
    """Controla el flujo del programa"""
    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")
        print("1. Convertir grados Centígrados a Fahrenheit")
        print("2. Convertir grados Fahrenheit a Centígrados")
        print("3. Salir")

        opcion = input("Selecciona una opción (1-3): ")

        if opcion == '1':
            convertirCF()
        elif opcion == '2':
            convertirFC()
        elif opcion == '3':
            print("Saliendo del programa... ¡Hasta luego!")
            print("cristopher Lopez = 33")
            break
        else:
            print("Opción no válida, intenta de nuevo.")

# Punto de entrada para ejecutar el programa
if __name__ == "__main__":
    menu()
```
